```python
import logging

import time

from contextlib import suppress

from train import RLDropping
import torch
import torch.nn as nn
import torch.nn.parallel

from timm.data import create_dataset

from timm.models import  load_checkpoint

from timm.data.transforms_factory import create_transform

try:
    from apex import amp

    has_apex = True
except ImportError:
    has_apex = False

from timm.models.vision_transformer import VisionTransformer

class VisionTransformerEval(VisionTransformer):
    def forward(self, x: torch.Tensor, thinking_stages=None) -> torch.Tensor:

        if len(thinking_stages) == 2:
            p = self.p = int(thinking_stages[0]) / 12
            x_out_stage_0 = self.forward_features(x, p)
            logits_stage_0 = self.forward_head(x_out_stage_0)
            # entropy_stage_0 = self.entropy(logits_stage_0)

            p = self.p = int(thinking_stages[1]) / 12
            x_out_stage_1 = self.forward_features(x, p, x_out_stage_0, thinking_stages)
            logits_stage_1 = self.forward_head(x_out_stage_1)

            # mask = torch.where(entropy_stage_0 < threshold, 0, 1)
            # output_logits = torch.where(entropy_stage_0 < threshold, logits_stage_0, logits_stage_1)
            return logits_stage_0, logits_stage_1


        if len(thinking_stages) == 3:
            p = self.p = int(thinking_stages[0]) / 12
            x_out_stage_0 = self.forward_features(x, p)
            logits_stage_0 = self.forward_head(x_out_stage_0)
            # entropy_stage_0 = self.entropy(logits_stage_0)

            p = self.p = int(thinking_stages[1]) / 12
            x_out_stage_1 = self.forward_features(x, p, x_out_stage_0, thinking_stages)
            logits_stage_1 = self.forward_head(x_out_stage_1)
            # entropy_stage_1 = self.entropy(logits_stage_1)

            p = self.p = int(thinking_stages[2]) / 12
            x_out_stage_2 = self.forward_features(x, p, x_out_stage_1, thinking_stages)
            logits_stage_2 = self.forward_head(x_out_stage_2)

            # Masks
            # mask_stage_0 = entropy_stage_0 < threshold
            # mask_stage_1 = ~mask_stage_0 & (entropy_stage_1 < threshold)

            # output_logits = logits_stage_2.clone()
            # output_logits = torch.where(mask_stage_0.repeat(1, 1000), logits_stage_0, output_logits)
            # output_logits = torch.where(mask_stage_1.repeat(1, 1000), logits_stage_1, output_logits)

            # mask = torch.where(
            #     mask_stage_0,
            #     0,
            #     torch.where(
            #         mask_stage_1,
            #         1,
            #         2
            #     )
            # )
            return logits_stage_0, logits_stage_1, logits_stage_2
_logger = logging.getLogger('validate')
def validate_thinking(model_type, pretrained_model, root_dir, num_classes, thinking_stages=[3, 6], n_heads=12,
             patch_size=16,batch_size=32,use_ema=True,use_amp='apex'):
    if torch.cuda.is_available():
        torch.backends.cuda.matmul.allow_tf32 = True
        torch.backends.cudnn.benchmark = True

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # create model
    in_chans = 3
    kwargs_model = {'depth': 12, 'drop_path_rate': 0.1, 'drop_rate': 0.0, 'embed_dim': 768, 'img_size': (224, 224),
                    'in_chans': in_chans,
                    'num_classes': num_classes, 'num_heads': n_heads, 'patch_size': patch_size}

    model = VisionTransformerEval(**kwargs_model)

    model.alpha = nn.ParameterList([nn.Parameter(torch.zeros(1)) for _ in range(12)])
    model.proj_layer = nn.Linear(64 * int(thinking_stages[0]), 768).to(device)

    if len(thinking_stages) == 3:
        model.proj_layer2 = nn.Linear(64 * int(thinking_stages[1]), 768).to(device)

    load_checkpoint(model, pretrained_model, use_ema)

    param_count = sum([m.numel() for m in model.parameters()])
    _logger.info('Model %s created, param count: %d' % (model_type, param_count))

    model = model.to(device)

    if use_amp == 'apex':
        model = amp.initialize(model, opt_level='O1')

    # criterion = nn.CrossEntropyLoss().to(device)

    dataset = create_dataset(
        root=root_dir,
        name='',
        split='validation',
    )


    model.eval()
    # ----**----#
    RLDropping(model, used_heads=n_heads, max_heads=12, ema=False)
    # ----**----#
    amp_autocast = suppress
    dataset.transform = create_transform((batch_size,in_chans) + tuple((224, 224)),
                                          no_aug=True)
    with torch.no_grad():
        # warmup, reduce variability of first batch time, especially for comparing torchscript vs non
        input = torch.randn((batch_size,in_chans) + tuple((224, 224))).to(device)
        with amp_autocast():
            model(input, thinking_stages=thinking_stages)

        end = time.time()
        for batch_idx, (input, target) in enumerate(dataset):

            # compute output
            input = input.to(device)
            with amp_autocast():
                logits0, logits1 = model(input[None], thinking_stages=thinking_stages)

                if len(thinking_stages) == 3:
                    logits2=  logits1[1]
                    logits1= logits1[0]

            break
                ### Code here


def main():
    validate_thinking('deit_base_patch16_224',
                      'output_data/20251024-014909-deit_base_patch16_224-224/last.pth.tar',
                      root_dir='../datasets/mini_imagenet',
                      num_classes=4,
                      thinking_stages=[3, 6],
                      n_heads=12,
                      use_amp=has_apex)

main()
```
